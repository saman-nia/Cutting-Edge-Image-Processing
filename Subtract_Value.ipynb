import the requirement libraries

```python
from PIL import Image, ImageFilter
import numpy as np
import random
import matplotlib.pyplot as plt
from random import randint
```

Import the Image with the "L" format that indicated the range of Image is 0 to 255 with the grayscale format.

```python
img = Image.open('../The-first-assignment-for-IPA/Lab1-Images/Lab1-Images/lena1.tif').convert('L')  # convert image to 8-bit grayscale
```

Store the position of each pixel.

```python
width, height = img.size
```

Define a fanction to add a random value with range of 0 to 255 and clip to min and max with the range of 0 to 255.

```python
def add_noise(a):
    return min(max(0, a + randint(0, 255)), 255)
```

Using a loop system to find and implement the function on that.

```python
for x in range (0, width):
    for y in range (0, height):
        a = img.getpixel((x,y))
        img.putpixel(((x,y)), add_noise(a))
        b = np.array(a)
```

Save the image in local memory.

```python
img.save('../The-first-assignment-for-IPA/Results/Random_Value.png')
```

Show the Minimum and Maximum pixel after produce the image.

```python
smal = np.amin(img)
big = np.amax(img)
```

```python
print ("The smallest value of pixel is = "), (smal), ("\nThe biggest value of pixel is = "), (big)
```
